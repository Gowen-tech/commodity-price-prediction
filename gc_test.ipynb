```python
import os
rootdir = "棕榈油/" # 输入相应的比对文件路径
files = sorted(os.listdir(rootdir))
files = [x for x in files if len(re.compile("比对").findall(x)) != 0]

for f in files:
    print("---------------------------------------")
    print(f)
    tmp = {}
    df1 = pd.read_csv(rootdir+f, usecols=[1,2])
    df2 = pd.read_csv(rootdir+f, usecols=[0,1])
    df3 = pd.read_csv(rootdir+f, usecols=[0,2])
    df2 = pd.merge(df3, df2, on="date").iloc[:, 1:]
    rst1 = ts.grangercausalitytests(df1, maxlag=3, addconst=True, verbose=True)
    rst2 = ts.grangercausalitytests(df2, maxlag=3, addconst=True, verbose=True)
```
